#### Voir les frames doublons

```python
import pandas as pd

# Suppose que df est déjà défini
cols_to_check = ['text_of_frame', 'username', 'servername']

# 1. Trouver les lignes dupliquées (hors colonne 'frame')
duplicated_mask = df.duplicated(subset=cols_to_check, keep=False)
duplicates_df = df[duplicated_mask].copy()

# 2. Grouper les frames identiques
grouped = (
    duplicates_df
    .groupby(cols_to_check)['frame']
    .apply(list)
    .reset_index()
)

# 3. Construire un seul message par groupe (en gardant le premier frame comme référence)
result = []
total_doublons = 0

for frames in grouped['frame']:
    if len(frames) > 1:
        first = frames[0]
        rest = frames[1:]
        result.append(f"Le frame {first} a pour doublons les frames {rest}")
        total_doublons += len(rest)  # Ne compter que les doublons, pas la première apparition

# 4. Affichage
for ligne in result:
    print(ligne)

# 5. Affichage du total
print(f"\nNombre total de doublons : {total_doublons}")
```

### **Phase 4 : Extraction de commandes systèmes et SQL sur chaque frame** 

```python
import re
import pandas as pd

# === Prompts
SQL_PROMPTS = [
    r"SQL>",                 # Oracle
    r"[a-zA-Z0-9_\-]+=#",    # PostgreSQL superuser
    r"[a-zA-Z0-9_\-]+=>",    # PostgreSQL normal user
]
SQL_PROMPT_REGEX = re.compile(rf"^({'|'.join(SQL_PROMPTS)})\s*(.*)", re.IGNORECASE)
SYSTEM_PROMPT_REGEX = re.compile(r"\[([^\s@]+)@([^\s\]]+)\s~\](\$|#)")

SQL_SIMPLE_COMMANDS = {"exit", "quit", "connect", "conn", "disconnect", "disc", r"\q"}


def extract_commands_with_type(text):
    if not isinstance(text, str):
        return []

    lines = text.splitlines()
    commands = []

    current_cmd = ""
    inside_sql = False

    for line in lines:
        line = line.rstrip()

        # === SYSTEM PROMPT
        sys_match = SYSTEM_PROMPT_REGEX.match(line)
        if sys_match:
            # Annuler toute commande SQL en cours (si elle n'est pas terminée par ;)
            if inside_sql:
                current_cmd = ""
                inside_sql = False

            cmd = line[sys_match.end():].strip()
            if cmd:
                commands.append((cmd, "sys"))
            continue

        # === SQL PROMPT
        sql_match = SQL_PROMPT_REGEX.match(line)
        if sql_match:
            # Si on était en train d’accumuler une commande SQL et qu’elle n’était pas terminée → on l’abandonne
            if inside_sql:
                current_cmd = ""
                inside_sql = False

            sql_line = sql_match.group(2).strip()

            # Si la commande est simple (exit, etc.) → extraire immédiatement
            if sql_line.lower().rstrip(";") in SQL_SIMPLE_COMMANDS:
                commands.append((sql_line, "sql"))
                continue

            # Démarrer une nouvelle commande SQL multi-ligne
            current_cmd = sql_line + "\n"
            inside_sql = True
            continue

        # === Accumuler lignes SQL jusqu’à trouver un ;
        if inside_sql:
            current_cmd += line.strip() + "\n"
            if line.strip().endswith(";"):
                commands.append((current_cmd.strip(), "sql"))
                current_cmd = ""
                inside_sql = False
            continue

        # === Autres lignes ignorées (hors SQL ou système)
        continue

    # On ne garde pas de commande SQL non terminée par ;
    return commands


# === Application sur df_cleaned avec duplication dynamique
def enrich_df_with_commands(df):
    all_rows = []

    for _, row in df.iterrows():
        command_tuples = extract_commands_with_type(row['text_of_frame'])

        if not command_tuples:
            continue

        historique_cmd = '\n\n'.join(cmd for cmd, _ in command_tuples)

        for cmd, cmd_type in command_tuples:
            new_row = row.copy()
            new_row['historique_cmd'] = historique_cmd
            new_row['cmd_sys'] = cmd if cmd_type == "sys" else ""
            new_row['cmd_sql'] = cmd if cmd_type == "sql" else ""
            all_rows.append(new_row)

    return pd.DataFrame(all_rows)


# === Exécution
df_cleaned = enrich_df_with_commands(df_cleaned)
```
